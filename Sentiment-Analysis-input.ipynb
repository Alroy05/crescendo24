```python
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm

sia = SentimentIntensityAnalyzer()
```

```python
input = "Put this product in your ass"
res = sia.polarity_scores(input)
print(res)
```

```python
# Define thresholds
neutral_threshold_low = -0.33
neutral_threshold_high = 0.33
sentiment_classifications=[]


compound_score = res['compound']
category = ''
if compound_score < neutral_threshold_low:
  category = 'Negative'
elif compound_score > neutral_threshold_high:
  category = 'Positive'
else:
  category = 'Neutral'
```
